```python
header = { # 헤더 딕셔너리 형태
    "Authorization": "KakaoAK {}".format(KAKAO_API_KEY) # 웬만하면 ctrl c + ctrl v 해라
}
header
```

이제는 파이썬 프로그램으로 요청해야함<br>
urllib, requests : 모듈 사용

```python
import requests

response = requests.get( # get 방식으로 요청하라고 했으니 get
    KAKAO_WEB_SEARCH_URL.format("디아블로4"),
    headers = header # 원래 헤더에는 많은 것이 들어있다. 그런데 거기에 인가정보만 +a로 넣어준 것 뿐이다.
)
response
```

```python
response.status_code
```

응답 코드
- 1xx : 처리중
- 2xx : success
- 3xx : redirection -> 추가 동작이 필요하다, 시간이 답이다.
- 4xx : client error -> 클라이언트에서 오류 발생
  - 404 : 잘못된 url로 요청함
- 5xx : server error -> 서버에서 오류 발생

response.content : html & css & js 같은 페이지 코드를 가져올 때 사용
response.json() : 시리얼라이즈 된 json 데이터를 자동으로 디시리얼라이즈 해줌

```python
datas = response.json()
datas
```

```python
datas.keys() # Json과 Dict가 비슷해서 가능
```

```python
documents = datas['documents'] # json 데이터를 여러 개 나열한 json array
documents # json array
```

```python
# json array를 dataframe으로 만들기
import pandas as pd
df = pd.DataFrame(documents)
df
```
