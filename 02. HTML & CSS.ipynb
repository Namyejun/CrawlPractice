```python
html_str = """
<html>
  <head>
    <title>안녕하세요</title>
  </head>
  <body>
    <div id="container">
      <p class='p1'>hello</p>
      <p>Bye</p>
    </div>
  </body>
</html>"""
```

- html 형식의 문자열 내에 원하는 데이터가 존재
- HTML Element에 원하는 데이터가 있다. -> dom 구조를 만든다.
- 즉 Element 객체를 만들자.

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html_str, 'html.parser')
soup
```

```python
type(soup)
```

# find
- find("태그명", "속성값을 딕셔너리로 표현") : 한 개의 엘리먼트 찾기
- find_all("태그명", "속성값을 딕셔너리로 표현") : 여러 개의 엘리먼트 찾기

```python
container_div = soup.find("div", {"id":"container"})
container_div
```

```python
type(container_div)
```

```python
soup.find_all("p")
```

```python
soup.find("p")
```

```python
soup.find_all("p",{"class":"p1"})
```

텍스트 추출
- 반드시 element 객체인 상태에서만 가능하다. find로 나온 결과나, find_all의 요소만  가능하다.

```python
soup.find("p",{"class":"p1"}).text
```

```python
for i in soup.find_all("p"):
  print(i.text)
```

```python
container_div = soup.find("div", {"id": "container"})
container_div
```

```python
container_div.find_all("p")
```

# 선택자(Selector)를 사용해서 찾기
- select("선택자") : 선택자에 의해 엘리먼트를 여러 개 선택
- select_one("선택자") : 선택자에 의해 엘리먼트를 한 개만 선택

```python
soup.select_one("#container")
```

```python
container_div.select_one(".p1")
```

```python
soup.select_one("#container > .p1")
```

```python
soup.select("#container > p")
```

# 텍스트, 속성 추출

```python
soup.select("#container > p")[1].text
```

```python
soup.find_all("p").text()
```

## .p1을 가져오고 싶다면?

```python
soup.select_one("#container > .p1").get("class") # get(속성명)
```
