셀레니움 사용을 고려하는 상황
- 비동기 통신에 의한 페이지 로딩이 힘든 경우
    - html에서 데이터를 가지고 와야하는데 안나와
    - fetch/xhr 확인을 해봤는데 API도 없어(종종 이런 경우 있음)
- 특정 페이지의 일부분을 캡쳐를 한다거나 파일을 다운로드 받아서 데이터를 수집한다거나 등등 사용자의 액션에 의한 데이터 수집이 필요한 경우

```python
YOUTUBE_ENDGAME_VIDEO_URL = "https://www.youtube.com/watch?v=Ko2NWhXI9e8&t=128s"
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
```

```python
def get_chrome_driver():
    # 1. 브라우저 옵션 세팅
    chrome_options = webdriver.ChromeOptions()
    
    # 2. driver 생성
    driver = webdriver.Chrome(
        service = Service(ChromeDriverManager().install()),
        options = chrome_options
    )
    return driver
```

```python
driver = get_chrome_driver()
```

```python
driver.get(YOUTUBE_ENDGAME_VIDEO_URL)
```

```python
# 스크린샷 저장
driver.save_screenshot("endgame_128s_screenshot.png")
driver.quit()
```

```python
# element만 캡처하기
from selenium.webdriver.common.by import By

driver = get_chrome_driver()
driver.get(YOUTUBE_ENDGAME_VIDEO_URL)

movie_player = driver.find_element(
    By.CSS_SELECTOR,
    "#movie_player"
)

movie_player
```

```python
# 영상 가운데 큰 재생버튼 눌러주기
try:
    driver.find_element(
    By.CSS_SELECTOR,
        ".ytp-large-play-button.ytp-button"
).click()
except:
    pass
```

```python
# element의 위치 찾기 (좌상단 좌표)
location = movie_player.location
location
# 캡처를 위해 우하단 좌표를 구해야함
```

```python
size = movie_player.size
size
```

```python
left = location['x']
top = location['y']
right = left + size["width"]
bottom = top + size["height"]
area = (left, top, right, bottom)
area
```

Youtube 영상 다운로드

```python
def download(title, link):
    response = requests.get(link, stream=True)
    save_path, size = "{}".format(title), 0
    with open(save_path, "wb") as f:
        for chunk in response.iter_content(chunk_size=1024):
            if chunk:
                size += 1024
                f.write(chunk)
    return size
```

```python
# 유튜브에서 영상을 다운 받으려면 www.youtube.com -> ssyoutube.com
# https://www.youtube.com/watch?v=Ko2NWhXI9e8&t=128s
# https://ssyoutube.com/watch?v=Ko2NWhXI9e8&t=128s
YOUTUBE_MOVIE_DOWNLOAD_URL = "https://ssyoutube.com/watch?v=Ko2NWhXI9e8&t=128s"

driver = get_chrome_driver()
driver.get(YOUTUBE_MOVIE_DOWNLOAD_URL)
```

```python
download_button = driver.find_element(
    By.CSS_SELECTOR,
    "#sf_result > div > div > div.info-box > div.link-box > div.def-btn-box > a"
)
download_button
```

```python
# 셀레니움을 통한 attribute 수집
title = download_button.get_attribute("download")
download_src = download_button.get_attribute("href")
title, download_src
```

```python
import requests
size = download(title, download_src)
```

```python
print("{:.3f}Mb".format(size/1024/1024))
```
