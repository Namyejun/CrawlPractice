```python
page = 1
```

```python
NAVER_SERIES_URL = f"https://series.naver.com/novel/categoryProductList.series?categoryTypeCode=all&page={page}"
```

```python
import requests
reponse = requests.get(NAVER_SERIES_URL)
```

```python
html_code = reponse.content #.decode(encoding='UTF-8', errors='strict')
```

```python
from bs4 import BeautifulSoup as BS
soup = BS(html_code, "html.parser")
soup
```

```python
items = soup.select("div#content > div.lst_thum_wrap > ul.lst_list > li > div.cont.cont_v2 > h3 > a")
items
```

```python
for item in items:
    title = item['title']
    href = "https://series.naver.com" + item['href']
    print(title)
    print(href)
```
