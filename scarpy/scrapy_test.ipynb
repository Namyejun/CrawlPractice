```python
import scrapy
import requests
from scrapy.http import TextResponse # XPATH 연습용
```

# XPath
- CSS Selector를 확장시킨 문법

## xpath 사용 방법

//*[@id="wrapSearch"]/div[3]/ul[3]/li[1]/a

body #wrapSearch > div:nth-child(3) > ul:nth-child(3) > li:nth-child(1) > a

### xpath 구성 요소
- // : 가장 상위 엘리먼트
- \* : 조건에 맞는 하위 엘리먼트를 모두 살펴봄 -> 이거 조금 이상함
  - div .txt
  - 후손 선택자의 역할
- [@id="wrapSearch"] : 조건 : id가 wrapSearch인 엘리먼트
- / : 바로 아래 엘리먼트를 살펴봄
  - div > .txt
  - 자식 선택자의 역할
- div[3] : div 태그에서 3번째 엘리먼트를 선택
  - 1부터 시작 

```python
response = requests.get("http://corners.gmarket.co.kr/Bestsellers")
response.status_code
```

```python
response.text
```

```python
text_response = TextResponse(
    response.url, # http://corners.gmarket.co.kr/Bestsellers
    body = response.text,
    encoding = 'utf-8'
) # arg 중요
text_response
```

```python
text_response.xpath('//*[@id="itemidx1"]/a') # 따옴표 무조건 작은 따옴표 id 떄문
```

```python
text_response.xpath('//*[@id="itemidx1"]/a/text()') # 데이터 부분에 태그가 들어있었는데 텍스트가 들어감
```

```python
# 텍스트 가져오기
text_response.xpath('//*[@id="itemidx1"]/a/text()').extract() # 데이터만 가져옴
```

```python
text_response.xpath('//*[@id="gBestWrap"]/div/div[3]/div/ul/li/a/text()').extract() # 오우야 맛좋다
```

```python
text_response.xpath('//*[@id="gBestWrap"]/div/div[3]/div/ul/li[1]/a/text()').extract()
```

```python
text_response.xpath('//*[@id="gBestWrap"]/div/div[3]/div/ul/li/a/text()').extract()[0]
```

```python
# 속성 추출은 속성명 앞에 @ 붙이기
text_response.xpath('//*[@id="gBestWrap"]/div/div[3]/div/ul/li/a/@href').extract()
```

```python
# 이미지 가져오기
text_response.xpath('//*[@id="gBestWrap"]/div/div[3]/div/ul/li/div/a/img/@data-original').extract()
```

```python
# 이미지 가져오기 후손 선택
text_response.xpath('//*[@id="gBestWrap"]/div/div[3]/div/ul/li//img/@data-original').extract()
```

```python
# 이미지 가져오기
text_response.xpath('//img/@data-original').extract()
```

```python
# 후손 선택자 실험 1 얘네 두개가 다름
text_response.xpath('//a/@href').extract()
```

```python
# 후손 선택자 실험 2 얘네 두개가 다름 따라서 // 이게 후손 선택자라고 사려됨.
text_response.xpath('//*[@id="gBestWrap"]/div/div[3]/div/ul/li//a/@href').extract()
```
